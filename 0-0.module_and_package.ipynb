### 사용 방법

#### 첫번째 방법

```python
import 모듈_이름

모듈_이름.모듈_함수()
a = 모듈_이름.모듈_함수()
```

````python
import calendar

calendar.prmonth(2025, 9)
````

#### 두번째 방법

```python
from 모듈_이름 import 모듈_함수

모듈_함수()
a = 모듈_함수()
```


````python
del calendar
calendar.prmonth(2025, 9)
````

````python
from calendar import prmonth
prmonth(2025, 9)
````

#### 두번째 방법 확장1

```python
from 모듈_이름 import 모듈_함수1, 모듈_함수2
```

````python
from calendar import prmonth, isleap
prmonth(2024, 2)
y2024_is_leap = isleap(2024)

prmonth(2025, 2)
y2025_is_leap = isleap(2025)

print(f'\n\n2024 is leap year? {y2024_is_leap}')
print(f'2025 is leap year? {y2025_is_leap}')
````

#### 두번째 방법 확장2

```python
from 모듈_이름 import *
```

````python
del prmonth
del isleap
prmonth(2024, 2)
````

````python
from calendar import *
prmonth(2024, 2)
y2024_is_leap = isleap(2024)

prmonth(2025, 2)
y2025_is_leap = isleap(2025)

print(f'\n\n2024 is leap year? {y2024_is_leap}')
print(f'2025 is leap year? {y2025_is_leap}')
````
