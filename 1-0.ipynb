## Colab
### Colab이란?

*  구글이 제공하는 클라우드 기반 Jupyter Notebook 환경
*  별도의 환경설정 없이 머신 러닝/딥러닝 사용 가능
*  GPU를 손쉽게 사용할 수 있음

    - GPU: 수천 개의 작은 코어 --> 반복 계산
    - ML/DL: 많은 양의 행렬,벡터 연산이 필요하므로 병렬처리 구조의 GPU가 빠름!

## Git/GitHub

### Git이란?
*  코드의 버전 관리 도구
*  GitHub: Git 저장소를 온라인에 올려 협업/백업하는 서비스

#### 주요 용어
*  Repository: 프로젝트 폴더 하나 = 코드/이력 모음
*  Remote: 원격 저장소
*  clone: 원격 저장소 --> 내 환경(로컬 저장소)으로 복사
*  commit: 변경 사항 저장
*  push: 로컬 저장소의 변경 사항을 온라인 저장소로 업로드, 반드시 commit이후 push 가능!
*  pull: 원격 저장소에서 변경된 내용을 로컬 저장소로 다운로드
*  pull request(PR): 협업 시 코드 수정 반영을 위한 요청

### Let's start Git!
1. GitHub 로그인
2. New repository
3. Settings --> Developer settings --> Personal access tokens --> Tokens (classic) --> Generate new token (classic)
4. repo 권한 체크 --> 생성 --> 표시된 토큰을 복사 (다시 볼 수 없으니 다른 곳에 저장 필수!)
5. Git 사용자 정보 설정
```bash
git config --global user.name "Your Name"
git config --global user.email "your_email@example.com"
```
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;6-1. 공개 레포지토리 clone: 코드를 가져와서 혼자 사용하고 싶을 때
```bash
# 1) GitHub에서 레포 생성

# 2) 공개 레포 복제
git clone https://github.com/Hyelim-Shin/AIdev.git

# 3) 복제한 폴더로 이동
cd /path/to/AIdev

# 4) 원격 재설정
git remote remove origin
git remote add origin https://github.com/{your_name}/AIdev.git
git push --set-upstream origin master
Username for: {your_name}
Password for: {your_token}
```

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;6-2. 공개 레포지토리 fork: 코드에 대해서 기여하고 싶을 때/협업
```bash
# 1) 공개 레포 포크

# 2) 내 레포 복제
git clone https://github.com/{your_name}/AIdev.git

# 3) 복제한 폴더로 이동
cd /path/to/AIdev

# 4) 원격 재설정
git push

# 5) (Additional!) upstream set
git remote add upstream https://github.com/Hyelim-Shin/AIdev.git
```

7. 다른 원격 브랜치로 이동
```bash
git fetch origin     # 원격 최신 브랜치 목록 가져오기
git branch -r        # 원격(remote) 브랜치 목록 보기
git checkout -b 2025 origin/2025
```
